```python
import sys
sys.path.append('../src/')

import cv2
import matplotlib.pyplot as plt
import numpy as np

from model import augment
from config import DATA_PATH
```

# Demonstrate image augmentation

Retrieve an image from the training set, run augmentation layer from model file and show before/after.

```python
fig = plt.figure(figsize = (8,4))

orig_img = cv2.imread(DATA_PATH + "Training/Fire/resized_frame13094.jpg")
aug_img = np.squeeze(augment(np.array([orig_img])).numpy())

plt.subplot(1, 2, 1)
plt.axis('off')
plt.imshow(cv2.cvtColor(orig_img, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.axis('off')
plt.imshow(cv2.cvtColor(aug_img, cv2.COLOR_BGR2RGB))

plt.show()
```
